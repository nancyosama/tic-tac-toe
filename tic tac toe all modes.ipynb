```python
from random import randint
```

```python
class human1:
    def __init__(self):
        print("[1] Human Vs. Human")
        print("[2] Human Vs. Computer")
        print("[3] Human Vs Smart Computer")
        print("[4] Smart Computer Vs Smart Computer")
        self.mode=int(input("please enter your mode"))
        self.board=[" "," "," "," "," "," "," "," "," "," "]
        self.check=False
        self.flag=0
        self.enable=((1,2,3),(4,5,6),(7,8,9),(1,4,7),(2,5,8),(3,6,9),(1,5,9),(3,5,7))
    def display_printing(self,board):
        print(self.board[1]+"|"+self.board[2]+"|"+self.board[3])
        print("-+-+-")
        print(self.board[4]+"|"+self.board[5]+"|"+self.board[6])
        print("-+-+-")
        print(self.board[7]+"|"+self.board[8]+"|"+self.board[9])
        print("--------------------------------")
    def pro(self):
        if (self.mode ==1):
            self.display_printing(" ")
            self.flag=1
        elif (self.mode==2):
            self.display_printing(" ")
            self.flag=2
        elif (self.mode==3):
            self.display_printing(" ")
            self.flag=3
        elif (self.mode==4):
            self.flag=4
            self.display_printing(" ")
    def select_x(self):
        if self.flag==1:
            for x in range(9):
                if self.check==True:
                    break
                print("HUMAN, X Turn")
                position=int(input("please enter your position:"))
                while self.board[position] !=" ":
                    print("please enter a valid position")
                    print("X Turn")
                    position=int(input("please enter your position:"))
                if self.board[position]==" ":
                    self.board[position]="X"
                    self.display_printing(self.board[position])
                for z in range(len(self.enable)):
                    if self.board[self.enable[z][0]]==self.board[self.enable[z][1]]==self.board[self.enable[z][2]]=="X":
                        print("X IS WINNING")
                        break
                else:
                    self.select_O()
                self.check=True
    def select_w(self):
        if self.flag==2:
            for y in range(9):
                if self.check==True:
                    break
                else:
                    print("COMPUTER, X Turn")
                    position=randint(1,9)
                    while self.board[position] !=" ":
                        print("please enter a valid position")
                        position=randint(1,9)
                    if self.board[position]==" ":
                        self.board[position]="X"
                        self.display_printing(self.board[position])
                    for z in range(len(self.enable)):
                        if self.board[self.enable[z][0]]==self.board[self.enable[z][1]]==self.board[self.enable[z][2]]=="X":
                            print("X IS WINNING")
                            break
                    else:   
                        self.select_O()
                    self.check=True
    def select_SC(self):
        if self.flag==3:
            for y in range(9):
                if self.check==True:
                    break
                else:
                    print("COMPUTER, X Turn")
                    for i in range(len(self.enable)):
                        if self.enable[i][0]==self.enable[i][1] and self.board[self.enable[i][2]]==" ":
                            position=self.enable[i][2]
                        elif self.enable[i][1]==self.enable[i][2] and self.board[self.enable[i][0]]==" ":
                            position=self.enable[i][0]
                        else:
                            position=randint(1,9)
                    while self.board[position] !=" ":
                        position=randint(1,9)
                    if self.board[position]==" ":
                        self.board[position]="X"
                        self.display_printing(self.board[position])
                    for z in range(len(self.enable)):
                        if self.board[self.enable[z][0]]==self.board[self.enable[z][1]]==self.board[self.enable[z][2]]=="X":
                            print("X IS WINNING")
                            break
                    else:   
                        self.select_O()
                    self.check=True
    def select_SC2(self):
        if self.flag==4:
            for y in range(9):
                if self.check==True:
                    break
                else:
                    print("COMPUTER 1, X Turn")
                    for i in range(len(self.enable)):
                        if self.enable[i][0]==self.enable[i][1] and self.board[self.enable[i][2]]==" ":
                            position=self.enable[i][2]
                        elif self.enable[i][1]==self.enable[i][2] and self.board[self.enable[i][0]]==" ":
                            position=self.enable[i][0]
                        else:
                            position=randint(1,9)
                    while self.board[position] !=" ":
                        position=randint(1,9)
                    if self.board[position]==" ":
                        self.board[position]="X"
                        self.display_printing(self.board[position])
                    for z in range(len(self.enable)):
                        if self.board[self.enable[z][0]]==self.board[self.enable[z][1]]==self.board[self.enable[z][2]]=="X":
                            print("X IS WINNING")
                            break
                    else:   
                        self.select_SCO()
                    self.check=True
    def select_O(self):
        for y in range(9):
            if self.check==True:
                break
            else:
                print("HUMAN,O Turn")
                position=int(input("please enter your position:"))
                while self.board[position] !=" ":
                    print("please enter a valid position")
                    print("O Turn")
                    position=int(input("please enter your position:"))
                if self.board[position]==" ":
                    self.board[position]="O"
                    self.display_printing(self.board[position])
                if self.board[1]==self.board[2]==self.board[3]=="O":
                    print("O is winning")
                    break
                elif self.board[4]==self.board[5]==self.board[6]=="O":
                    print("O is winning")
                    break
                elif self.board[7]==self.board[8]==self.board[9]=="O":
                    print("O is winning")
                    break
                elif self.board[1]==self.board[4]==self.board[7]=="O":
                    print("O is winning")
                    break
                elif self.board[2]==self.board[5]==self.board[8]=="O":
                    print("O is winning")
                    break
                elif self.board[3]==self.board[6]==self.board[9]=="O":
                    print("O is winning")
                    break
                elif self.board[1]==self.board[5]==self.board[9]=="O":
                    print("O is winning")
                    break
                elif self.board[3]==self.board[5]==self.board[7]=="O":
                    print("O is winning")
                    break
                else:
                    if self.flag==1:
                        self.select_x()
                    elif self.flag==2:
                        self.select_w()
                    else:
                        self.select_SC()
                self.check=True
    def select_SCO(self):
        for y in range(9):
            if self.check==True:
                break
            else:
                print("COMPUTER 2, O Turn")
                for i in range(len(self.enable)):
                    if self.enable[i][0]==self.enable[i][1] and self.board[self.enable[i][2]]==" ":
                        position=self.enable[i][2]
                    elif self.enable[i][1]==self.enable[i][2] and self.board[self.enable[i][0]]==" ":
                        position=self.enable[i][0]
                    else:
                        position=randint(1,9)
                while self.board[position] !=" ":
                    position=randint(1,9)
                if self.board[position]==" ":
                    self.board[position]="O"
                    self.display_printing(self.board[position])
                for z in range(len(self.enable)):
                    if self.board[self.enable[z][0]]==self.board[self.enable[z][1]]==self.board[self.enable[z][2]]=="X":
                        print("O IS WINNING")
                        break
                    else:   
                        self.select_SC2()
                    self.check=True
```
